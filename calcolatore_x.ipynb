```python
import numpy as np
from scipy.stats import poisson

def calcola_probabilita_x(media_gol_casa, media_gol_trasferta):
    # Calcoliamo la probabilità dei risultati esatti fino a 4 gol
    prob_x = 0
    risultati = []

    for i in range(5): # 0, 1, 2, 3, 4 gol
        p_casa = poisson.pmf(i, media_gol_casa)
        p_trasferta = poisson.pmf(i, media_gol_trasferta)
        prob_x += p_casa * p_trasferta
        risultati.append(f"{i}-{i}: {p_casa * p_trasferta:.2%}")

    return prob_x, risultati

# --- TEST MANUALE ---
# Inserisci qui la media gol segnati/subiti delle due squadre
media_casa = 1.25  # Esempio: Squadra A segna/subisce 1.25 gol mediamente
media_trasferta = 1.10 # Esempio: Squadra B segna/subisce 1.10 gol mediamente

prob, dettagli = calcola_probabilita_x(media_casa, media_trasferta)

print(f"📊 Probabilità Totale Pareggio (X): {prob:.2%}")
print("Dettaglio risultati esatti:", dettagli)
```
